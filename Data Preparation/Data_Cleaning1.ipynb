```python
#Since we want to predict if a review is a popular review enough we need to find:
    #Books that have enough number of reader in order to assess popularity
    #Else we are unable to determine if the reviews are unpopular or merely unseen by the other users

#Thus, we have to clean the data from unpopular books.
#We will analyze a book's popularity based on their Combined_Likes_Count (total number of likes from all of the review in the book) and Num_Reviews (Total number of reviews a book receive)
```

```python
#Analysis of each book with their number of reviews
book_dtf_review_analysis = book_dtf[['Book_ID', 'Num_Review']].sort_values(by=['Book_ID'])
book_dtf_review_analysis.drop_duplicates(subset=['Book_ID'], inplace=True)
print(book_dtf_review_analysis)
```

```python
book_dtf_review_analysis.describe()
```

```python
#There's a huge std between each book's Num_Reviews
#To really make sure we have on ly books with sufficient amount of popularity, we will drop books with values less than Q3 of Num_Reviews

#GANTI
#Q3 of Num_Review is 32, so we will drop all the books with lesser than 32 total reviews
book_dtf = book_dtf[book_dtf.Num_Review >= book_dtf_review_analysis.Num_Review.quantile([0.75][0])]
print(book_dtf.shape)
```

```python
#Analysis of each book with their accumulated number of Likes from all the review
book_dtf_LikesComments_analysis = book_dtf[['Book_ID', 'Likes_Count']]
book_dtf_LikesComments_analysis = book_dtf_LikesComments_analysis.groupby("Book_ID")["Likes_Count"].sum().to_frame(name = "Combined_Likes_Count")
print(book_dtf_LikesComments_analysis)
print(book_dtf_LikesComments_analysis.shape)
```

```python
book_dtf_LikesComments_analysis.describe()
```

```python
#There's a huge std between each book's Combined_Likes_Count
#To really make sure we have only books with sufficient amount of popularity, we will drop books with values less than Q2 of Combined_Likes_Count 
#We use Q2 because the analysis is done on books which has been dropped based on their Num_Reviews, dropping them with Q3 of Combined_Likes_Count too will result in too much data loss

#Q2 of Combined_Likes_Count is 43, so we will drop books with lesser than 43 of Combined_Likes_Count
book_dtf_LikesComments_analysis = book_dtf_LikesComments_analysis[book_dtf_LikesComments_analysis.Combined_Likes_Count >= book_dtf_LikesComments_analysis.Combined_Likes_Count.quantile([0.5][0])]
book_dtf = book_dtf.merge(book_dtf_LikesComments_analysis,on='Book_ID')

#Drop the duplicates caused by the merge
book_dtf = book_dtf.reset_index(drop = True).drop_duplicates().sort_values(by=['Book_ID'])
print(book_dtf.shape)
```
